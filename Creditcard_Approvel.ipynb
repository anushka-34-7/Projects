```python
df=df.drop(['married','ethnicity'],axis=1)
```

```python
df=df.drop(['citizen','zipcode'],axis=1)
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
# model implementation

# target sepration

x=df.drop(columns='approved',axis=1)
y=df['approved']
```

```python
x.columns
```

```python
y.name
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
x.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# scaling-preprocessing

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
# LogisticRegression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train_scaled,y_train)
```

```python
# prediction
y_pread_train = model.predict(x_train_scaled)
y_pread_test = model.predict(x_test_scaled)
```

```python
# score
model.score(x_train_scaled,y_train)
```

```python
model.score(x_test_scaled,y_test)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pread_test)
cm
```

```python
# classification report
from sklearn.metrics import classification_report
cr = classification_report(y_test,y_pread_test)
print(cr)
```

```python
# implementing another model Decisioon tree

from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(x_train,y_train)
```

```python
# prediction

y_pred_train_dt =dt_model.predict(x_train)
y_pred_test_dt =dt_model.predict(x_test)
```

```python
# evaulation- accuracy

from sklearn.metrics import accuracy_score
print('training_accuracy:',accuracy_score(y_train,y_pred_train_dt))
print('Testing_accuracy:',accuracy_score(y_test,y_pred_test_dt))
```

```python
# confusion matrix
print("testing confusion matrix")
print(confusion_matrix(y_test,y_pred_test_dt))
```

```python
# classification report
print('classification report')
print(classification_report(y_test,y_pred_test_dt))
```

```python
# implement a new model randomforestclassifier
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42,n_estimators=30)
rf_model.fit(x_train,y_train)
```

```python
# predictions
y_pred_train_rf =rf_model.predict(x_train)
y_pred_test_rf =rf_model.predict(x_test)
```

```python
# accuracy
print('training_accuracy:',accuracy_score(y_train,y_pred_train_rf))
print('Testing_accuracy:',accuracy_score(y_test,y_pred_test_rf))
```

```python
# confusionm matrix
print('Testing confusion matrix')
print(confusion_matrix(y_test,y_pred_test_rf))
```

```python
#classification report

print('classification report')
print(classification_report(y_test,y_pred_test_rf))
```
