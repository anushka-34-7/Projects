```python
# train n rest split (80% data will use for train and 20% for test)
# scikit

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_train_prediction = lr.predict(x_train)
y_test_prediction = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

# the reason for using mae,mse,rmse,r2 is that we never judge a model by accuracy in linearregression , eare we used error based metrics

mae=mean_absolute_error(y_test,y_test_prediction)
mse=mean_squared_error(y_test,y_test_prediction)
rmse=np.sqrt(mse)
r2=r2_score(y_test,y_test_prediction)

print('MAE',mae)
print('MSE',mse)
print('RMSE',rmse)
print('R2',r2)
```

```python
mae=mean_absolute_error(y_train,y_train_prediction)
mse=mean_squared_error(y_train,y_train_prediction)
rmse=np.sqrt(mse)
r2=r2_score(y_train,y_train_prediction)

print('MAE',mae)
print('MSE',mse)
print('RMSE',rmse)
print('R2',r2)
```
